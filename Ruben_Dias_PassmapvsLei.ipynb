```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
df = pd.read_excel(r'D:\FOOTBALL ANALYTICS\EPL\RubenDiasvsLEI.xlsx')
```

```python
df.head(10)
```

```python
#convert the data to match the mplsoccer statsbomb pitch
df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#000009')
ax.patch.set_facecolor('#000009')

#this is how we create the pitch
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#000009', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

#Draw the pitch on the ax figure as well as invert the axis for this specific pitch
pitch.draw(ax=ax)
plt.gca().invert_yaxis()


# Plot the completed passes
lc1 = pitch.lines(df.x, df.y,
                  df.endX, df.endY,
                  lw=5, label='completed passes',
                  color='skyblue', ax=ax, transparent=True,comet=True)
# setup the legend
ax.legend(facecolor='#22312b', handlelength=5, edgecolor='None', fontsize=8, loc='upper left')

# Set the title
ax.set_title(f'Ruben Dias passes vs Leicester City', fontsize=20,c='white')
fig.text(.21,.14,f'Made by @RanojoySengupta / twitter',fontstyle='italic',fontsize=10,color='w')
```
